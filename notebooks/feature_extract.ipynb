```python
import pandas as pd
import pickle
import numpy as np
```

```python
df = pickle.load(open('dataset_level2.pkl','rb'))
```

```python
df

# batting_team
# bowling team
# city
# current score
# ball left
# wickets left
# current rr
# last five
```

```python
df.isnull().sum()
```

```python
df[df['city'].isnull()]['venue'].value_counts()
```

```python
cities = np.where(df['city'].isnull(),df['venue'].str.split().apply(lambda x:x[0]),df['city'])
```

```python
df['city'] = cities
df.isnull().sum()
```

```python
df.drop(columns=['venue'],inplace=True)
```

```python
df
```

```python
eligible_cities = df['city'].value_counts()[df['city'].value_counts() > 600].index.tolist()
```

```python
df = df[df['city'].isin(eligible_cities)]
```

```python
df
```

```python
df['current_score'] = df.groupby('match_id').cumsum()['runs']
```

```python
df
```

```python
df['over'] = df['ball'].apply(lambda x:str(x).split(".")[0])
df['ball_no'] = df['ball'].apply(lambda x:str(x).split(".")[1])
df
```

```python
df['balls_bowled'] = (df['over'].astype('int')*6) + df['ball_no'].astype('int')
df
```

```python
df['balls_left'] = 120 - df['balls_bowled']
df['balls_left'] = df['balls_left'].apply(lambda x:0 if x<0 else x)
df
```

```python
df['player_dismissed'] = df['player_dismissed'].apply(lambda x:0 if x=='0' else 1)
df['player_dismissed'] = df['player_dismissed'].astype('int')
df['player_dismissed'] = df.groupby('match_id').cumsum()['player_dismissed']
df['wickets_left'] = 10 - df['player_dismissed']
```

```python
df
```

```python
df['crr'] = (df['current_score']*6)/df['balls_bowled']
```

```python
df
```

```python
groups = df.groupby('match_id')

match_ids = df['match_id'].unique()
last_five = []
for id in match_ids:
    last_five.extend(groups.get_group(id).rolling(window=30).sum()['runs'].values.tolist())
```

```python
df['last_five'] = last_five
```

```python
df
```

```python
final_df = df.groupby('match_id').sum()['runs'].reset_index().merge(df,on='match_id')
```

```python
final_df=final_df[['batting_team','bowling_team','city','current_score','balls_left','wickets_left','crr','last_five','runs_x']]
```

```python
final_df.dropna(inplace=True)
```

```python
final_df.isnull().sum()
```

```python
final_df
```

```python
final_df = final_df.sample(final_df.shape[0])
```

```python
final_df.sample(2)    # shows random arrangement
```

```python
X = final_df.drop(columns=['runs_x'])
y = final_df['runs_x']
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=1)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from sklearn.metrics import r2_score,mean_absolute_error
from sklearn.model_selection import RandomizedSearchCV
from sklearn.metrics import accuracy_score
```

```python
trf = ColumnTransformer([
    ('trf',OneHotEncoder(sparse=False,drop='first'),['batting_team','bowling_team','city'])
]
,remainder='passthrough')
```

```python
scaler = StandardScaler()
```

```python
rf_regressor = RandomForestRegressor(n_estimators=100,min_samples_split=5,min_samples_leaf=1,max_depth=None,bootstrap=True)
```

```python
pipe = Pipeline(steps=[
    ('step1',trf),
    ('step2',scaler),
    #('step3',RandomForestRegressor(n_estimators=100,max_depth=None, min_samples_split=2, min_samples_leaf=1,max_features=1.0,n_jobs=None, random_state=None))
    #('step3',XGBRegressor(n_estimators=100,learning_rate=0.2,max_depth=5))
    ('step3',rf_regressor)
])
```

```python
param_dist = {
    'step3__n_estimators': [100, 200, 300],
    'step3__max_depth': [None, 10, 20, 30],
    'step3__min_samples_split': [2, 5, 10],
    'step3__min_samples_leaf': [1, 2, 4],
    'step3__bootstrap': [True, False]
}
```

```python
random_search = RandomizedSearchCV(estimator=pipe, param_distributions=param_dist,
                                   n_iter=10, scoring='neg_mean_squared_error', cv=5, n_jobs=-1)
```
